```python
import easyocr
from pdf2image import convert_from_path
import os

import numpy
```

```python
def convert_pdf_to_text(pdf_path, output_text_path):
    """
    Convert PDF to text format
    """
    try:
        # Convert PDF to images
        images = convert_from_path(pdf_path)
        
        # Initialize EasyOCR reader
        reader = easyocr.Reader(['en'])
        
        # Create text file
        with open(output_text_path, 'w', encoding='utf-8') as f:
            # Process each page
            for i, image in enumerate(images):
                # Extract text using EasyOCR
                results = reader.readtext(numpy.array(image))
                text = '\n'.join([result[1] for result in results])
                
                # Write text to file
                f.write(text + '\n')
                
                # Add separator between pages
                if i < len(images) - 1:
                    f.write('_' * 50 + '\n\n')
        
        print(f"Successfully converted PDF to text: {output_text_path}")
        return True
    except Exception as e:
        print(f"Error converting PDF to text: {str(e)}")
        return False
```

```python
input_folder = "/home/shaon/Handwritten-Form-Digitization-Portal/final_selected"
output_folder = "/home/shaon/Handwritten-Form-Digitization-Portal/final_selected_text"

os.makedirs(output_folder, exist_ok=True)

pdf_files = sorted([f for f in os.listdir(input_folder) if f.endswith('.pdf')])

for i, filename in enumerate(pdf_files, start=1):
    pdf_path = os.path.join(input_folder, filename)
    pdf_name = filename.split(".")[0]
    print(pdf_name)
    output_text_path = os.path.join(output_folder, pdf_name + ".txt")

    print(output_text_path)

    convert_pdf_to_text(pdf_path, output_text_path)


```
